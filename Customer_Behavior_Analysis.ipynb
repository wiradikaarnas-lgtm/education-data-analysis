```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

df.head()

df.groupby("day")["total_bill"].mean().plot(kind="bar")
plt.title("Average Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Average Total Bill")
plt.show()

df.groupby("time")["tip"].mean().plot(kind="bar")
plt.title("Average Tip by Time (Lunch vs Dinner)")
plt.show()


```

## Key Insights

1. Weekend (Saturday and Sunday) shows significantly higher average total bills compared to weekdays, which indicates a shift in customer behavior when they have more leisure time. This suggests restaurants could optimize staffing and inventory for weekends.

2. Dinner tips are higher than lunch tips, implying that customers place higher perceived value on evening dining experiences. This could justify premium pricing or upselling strategies during dinner hours.
